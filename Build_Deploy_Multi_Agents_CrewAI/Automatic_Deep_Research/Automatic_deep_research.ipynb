## Background: The Research Challenge

As a research consultant, you face a constant challenge:
- **High Volume**: Multiple research requests from demanding clients

## System Architecture: The Research Crew

### The Four-Agent Team

#### 👤 **Research Planner**
- **Role**: Strategic Orchestrator
- **Responsibility**: Analyzes the user's query and creates a research roadmap
- **Key Tasks**:
  - Break down complex questions into specific topics
  - Identify key research questions for each topic
  - Set success criteria for the research
- **Output**: Structured research plan with clear objectives

#### 🔍 **Internet Researcher**
- **Role**: Data Gatherer
- **Responsibility**: Collects information from multiple online sources
- **Tools Available**:
  - **EXA Search Tool**: Semantic web search for finding relevant articles
  - **Web Scraper Tool**: Extract and parse content from websites
- **Key Tasks**:
  - Execute research plan systematically
  - Gather diverse sources
  - Cite all sources with credibility notes
- **Output**: Comprehensive research data with citations

#### ✅ **Fact Checker**
- **Role**: Quality Assurance Officer
- **Responsibility**: Validates information accuracy and identifies gaps
- **Key Tasks**:
  - Cross-reference data across multiple sources
  - Identify conflicting information
  - Flag potential misinformation or biased sources
  - Highlight missing research areas
- **Output**: Quality assessment report with reliability ratings

#### 📝 **Report Writer**
- **Role**: Communication Expert
- **Responsibility**: Transforms research into professional reports
- **Key Tasks**:
  - Structure findings logically
  - Create executive summaries
  - Synthesize verified data into narrative
  - Format with proper citations and formatting
- **Output**: Publication-ready research report

### Create tool instances /Research Tools: Empowering the Agents

As you can see in the above workflow, the **Internet Researcher Agent** and the **Fact checker** have tools, so that they can better do their job. In particular, it has given to these two agents access to search the internet and scrape information from the retrieved webpages.

There are different tools inside CrewAI you can use to search the web, here it will be used the [**EXA Search Web Loader**](https://docs.crewai.com/en/tools/search-research/exasearchtool#exa-search-web-loader) tool, which is designed to perform a semantic search for a specified query from a text’s content across the internet. It utilizes the [exa.ai](https://exa.ai/) API to fetch and display the most relevant search results based on the query provided by the user. exa.ai enhances semantic search by capturing richer contextual relationships between concepts, allowing for more precise information retrieval than conventional embedding approaches.

For webscraping, it has been selected the [**Scrape Website**](https://docs.crewai.com/en/tools/web-scraping/scrapewebsitetool) tool, which is designed to extract and read the content of a specified website.

To be able to assign these tools to the agents, it is necesseary firstly to define instances of these tools. 

#### 🔍 EXA Search Tool

**Purpose**: Intelligent semantic web search

**Capabilities**:
- Searches the internet using natural language
- Returns highly relevant articles and content
- Provides source URLs and publication dates
- Semantic matching (finds relevant content even with different terminology)

````python
# import the tools instances
from crewai_tools import EXASearchTool, ScrapeWebsiteTool
````

````python
# Get the EXA_API_KEY
def setup_api_key(key_name: str) -> str:
    """Retrieve and validate API key from Colab secrets."""
    try:
        api_key = userdata.get(key_name)
        if not api_key:
            raise ValueError(f"{key_name} not found in Colab secrets")

        # Display masked key
        masked = api_key[:4] + "****" + api_key[-4:] if len(api_key) > 8 else "***"
        print(f"✓ {key_name} loaded successfully ({masked})")
        return api_key
    except Exception as e:
        print(f"❌ Error loading {key_name}: {e}")
        raise

exa_api_key = setup_api_key('EXA_API_KEY')
os.environ["EXA_API_KEY"] = exa_api_key
````

````python
# Create the EXASearchTool instance
exa_search_tool = EXASearchTool(base_url = "https://api.exa.ai")
````

## Agents section : Defining the Research Team

### Agent Parameters 

Each CrewAI agent is configured with these key parameters:

| Parameter | Purpose | Example |
|-----------|---------|---------|
| `role` | Agent's job title/function | "Research Planner" |
| `goal` | Specific outcome to achieve | "Break down queries into research topics" |
| `backstory` | Context about expertise (helps LLM roleplay) | "You are a research strategist..." |
| `tools` | List of tools available | `[exa_search_tool, scrape_website_tool]` |
| `verbose` | Show detailed execution logs | `True` (useful for debugging) |
| `max_rpm` | Maximum requests per minute (rate limiting) | `150` |
| `max_iter` | Maximum iterations before completion | `15` |


````python
# ===== AGENT DEFINITION =====
# Role: Describes the agent's job title and primary function
# Goal: The specific outcome the agent should achieve
# Backstory: Provides context about the agent's expertise (helps LLM roleplay)
# LLM: The language model powering the agent's reasoning
# Verbose: Shows detailed execution steps (useful for debugging)


# define the research planner agent
research_planner = Agent(
    role="Research Planner",
    goal="Analyze queries and break them down into smaller, specific research topics.",
    backstory=(
         "You are a research strategist who excels at breaking down complex questions "
         "into manageable research components. You identify what needs to be researched "
         "and create clear research objectives."
    ),
    verbose=True, # set to True to see detailed agent actions
    max_rpm=150,  # Allow 150 requests per minute
    max_iter=15   # Stop after 15 iterations max
)


# define the internet researcher agent
researcher = Agent(
    role="Internet Researcher",
    goal="Research thoroughly all assigned topics",
    backstory=(
        "You are an expert internet researcher skilled at gathering accurate "
        "and relevant information from diverse online sources. You utilize "
        "various research tools to find credible data efficiently."
    ),
    tools=[exa_search_tool, scrape_website_tool],
    verbose=True,
    max_rpm=150,
    max_iter=15
)


# define the fact checker agent
fact_checker = Agent(
    role="Fact Checker",
    goal=(
        "Verify data for accuracy, identify inconsistencies, "
        "and flag potential misinformation"
    ),
    backstory=(
        "You are a meticulous fact-checker dedicated to ensuring the accuracy "
        "and reliability of information. You cross-reference data from multiple "
        "credible sources to identify inconsistencies and verify facts."
    ),
    tools=[exa_search_tool, scrape_website_tool],
    verbose=True,
    max_rpm=150,
    max_iter=15
)

# define the report writer agent
report_writer = Agent(
    role="Report Writer",
    goal="Write clear, concise, and well-structured reports based on gathered information",
    backstory=(
        "You are a skilled report writer adept at synthesizing complex information "
        "into clear and concise documents. You organize data logically and present "
        "it in an engaging manner tailored to the target audience."
    ),
    verbose=True, # it allows to monitor the process
    max_rpm=150,
    max_iter=15
)
````

_Note_: Performance Tuning Tips. Adjusting Parameters such as:  
- **Increase max_iter:** If agents need more reasoning steps  
- **Decrease max_rpm:** If hitting API rate limits  
- **Set verbose=False:** For production (cleaner output)  
- **Adjust max_iter=5:** For simple queries (faster execution)

## Defining Work: Task Configuration

### Task Execution Order  

Tasks execute sequentially in the order specified:

1️⃣ Research Plan → 2️⃣ Gather Data → 3️⃣ Verify → 4️⃣ Write Report  

Each task receives output from the previous task as input.


#### Task 1: Create Research Plan

**Purpose**: Transform user query into actionable research plan      
**Expected output**: A research plan with main research topics to investigate, key questions for each topic, and success criteria for the research.    

#### Task 2: Gather Research Data  

**Purpose**: Execute research plan and collect information      
**Expected output**: Comprehensive research data including: information for each research topic, and citations used along with source credibility notes.      

#### Task 3: Verify Information Quality  

**Purpose:** Validate findings and ensure accuracy   
**Expected output**: A report with the all the collected data, and its review. It should include consistency check results and source reliability ratings.    

#### Task 4: Write Final Report  

**Purpose:** Synthesize verified findings into professional report  
**Expected output**: The final research report. In addition to the full answer, it should have an executive summary, and complete source citations.   

````python
# ===== TASK DEFINITION =====

#For each `Task` you need to define the following parameters:
#- `description`: A thorough description of the task. You can even break it down into different items.
#- `expected_output`: what should the output return. Be specific, specially if you want any structure in your result, like a dictionary with specific keys.
#- `agent`: who is performing the task? You need to match the task to one of the agents you already defined


# define the create research plan task
create_research_plan_task = Task(
    description=(
        "Based on the user's query, break it down into specific topics and key questions, "
        "and create a focused research plan."
        "The user's query is: {user_query}"
    ),
    expected_output=(
        "A research plan with main research topics to investigate, "
        "key questions for each topic, and success criteria for the research."
        ),
    agent=research_planner,
)

# define the gather research data task
gather_research_data_task = Task(
    description=(
        "Using the research plan, collect information on all identified topics. "
        "Cite all sources used."
    ),
    expected_output=(
        "A comprehensive research report containing all gathered information, "
        "citations, and sources."
    ),
    agent=researcher,
)

#define the verify information quality task
verify_information_quality_task = Task(
    description=(
        "Review all collected research. Identify any conflicting information, "
        "potential misinformation, or gaps that need addressing."
    ),
    expected_output=(
        "A detailed analysis of the research quality, highlighting any inconsistencies, "
        "misinformation, or missing data."
    ),
    agent=fact_checker,
)

# define the write final report task
write_final_report_task = Task(
    description=(
        "Create a comprehensive report that answers the original query using all verified research data. "
        "Structure it with clear sections, include citations, and provide actionable insights."
    ),
    expected_output=(
        "A well-structured final report that addresses the original query, "
        "including all relevant data, citations, and insights."
    ),
    agent=report_writer,
)
````

## Assembling the Crew: Orchestrating the Research Process  

### Creating the Crew

Once the agents and tasks have been defined, the crew can be created. In order to so, it is necessary set up the following arguments:
- `agents`: list of agents in the crew
- `tasks`: list of tasks in the crew. The tasks should be listed in the order they should be executed

````python
# create the crew with the defined agents and tasks
crew = Crew(
    agents=[
        research_planner, # Strategic orchestrator
        researcher,       # Data gatherer
        fact_checker,     # Quality assurance
        report_writer],   # Final synthesis
    tasks=[
        create_research_plan_task,    # Task 1
        gather_research_data_task,    # Task 2
        verify_information_quality_task,  # Task 3
        write_final_report_task           # Task 4
        ]
)
````

### Execute the Researcher Crew

Before executing the crew, you need to define the query, which will be used as input for the tasks.

````python
# Write your query, which will be used as input for the tasks.
#user_query="Summarize the impact of AI on the industry advancements in the last five years."
user_query="List top 3 AI platforms for developers and briefly describe their focus areas."
````

````python
# Execute the complete research workflow
result = crew.kickoff(
    inputs={
        "user_query": user_query,
    }
)

# Result structure:
# - output: Final research report
# - Used agents: All 4 agents in sequence
# - Execution time: Typically 5-15 minutes depending on topic
````

````python
# Print the final report 
from IPython.display import Markdown
Markdown(result.raw)
````

````python
# Save the report in .md format
with open('report.md', 'w') as f:
    f.write(result.raw)
print("Report saved to report.md")
````
